```python
# diff ways to plot
# 1st way:
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

## Matplotlib workflow

```python
# 0) Import matplotlib and prep it for usage
%matplotlib inline
import matplotlib.pyplot as plt

# 1) Prep Data
x = [1,2,4,5]
y = [11,23,34,55]

# 2) Setup Plot
fig, ax = plt.subplots(figsize = (5,5)) 

# 3) Plot data
ax.plot(x,y)


# 4) Customizing Plot
ax.set(title = "Example",
       xlabel = "Fuck Around"
      ,ylabel = "Find Out")
fig.set(label = "Example Figure")
plt.show()

# 5) Save and show
fig.savefig("../Images/exampleplot.png")
```

* When plotting something quickly use pyplot method
* When plotting something advanced, use OO method

```python
over50 = hDisease[hDisease["age"]>50]
over50.head()
```

```python
over50["age"].plot.hist()
```

```python
#Pyplot method
over50.plot(kind = 'scatter',
           y = "chol",
           x = "age",
           c = "target");
```

```python
plt.show()
```

```python
#OO Method
fig, ax = plt.subplots()
over50.plot(kind = 'scatter',
           x = "age",
           y = "chol",
           c = "target",
           ax = ax)
ax.set_xlim([45,90])
```

```python
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10,10),
                             sharex = True)
# AXIS 1
scatter = ax1.scatter(x = over50["age"],
           y = over50["chol"],
           c = over50["target"])
ax1.set_xlim([45,90])

# Customizing
ax1.set(title = "Title of the plot 1",
      ylabel = "Cholesterol");

# Legend
ax1.legend(*scatter.legend_elements(), title = "Target")

# Mean
average = np.mean(over50["chol"])

# Meanline
ax1.axhline(y = average, color = 'm', linestyle = "--", label = f'Average: {average}')

# AXIS 2
scatter2 = ax2.scatter(x = over50["age"],
           y = over50["thalach"],
           c = over50["target"])
ax2.set_xlim([45,90])

# Customizing
ax2.set(title = "Title of the plot 2",
      xlabel = "Age",
      ylabel = "Max Heart Rate");

# Legend
ax2.legend(*scatter.legend_elements(), title = "Target")

# Mean
average1 = np.mean(over50["thalach"])

# Meanline
ax2.axhline(y = average1, color = 'm', linestyle = "--", label = f'Average: {average1}')


# Figure Changes
fig.suptitle(t = "Title of the figure",fontsize = 15, fontweight="bold");
```
